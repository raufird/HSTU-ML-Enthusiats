```python
import numpy as np 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
```

```python
# set random seed for reproductibility

np.random.seed(42)
```

```python
# Generate random dataset with 2 features and 3 clusters

X = np.random.randn(150, 2)
X[:50] += 5
X[50:100] -= 5
y = np.concatenate([np.zeros(50), np.ones(50), np.ones(50)*2])
```

```python
# instantiate KMeans algorithm with 3 clusters

kmeans = KMeans(n_clusters=3)
```

```python
# Fit KMeans to data

kmeans.fit(X)
```

```python
# pridict cluster labels for data points

y_pred = kmeans.predict(X)
```

```python
# plot data pionts with different colors for each cluster

plt.scatter(X[:,0],X[:,1], c= y_pred)
plt.title("K-means Cluster")
plt.show()
```

```python
# Get cluster assignments for each data point

print(kmeans.labels_)
```

```python
# Get cluster centroids

print(kmeans.cluster_centers_)
```
